```python
import pandas as pd
from datetime import datetime, timedelta
import random, string
```

```python
# 读取学生信息
student_info = pd.read_excel('student_import_template.xlsx')
```

```python
# 生成学生信息
def generate_correct_stu_info():
    # 重新读取原始Excel文件
    student_df = pd.read_excel('student_import_template.xlsx')

    # 生成700行的空DataFrame，保持原有列名
    num_students = 700
    
    # 创建一个包含700行的新DataFrame，列名与原文件相同
    new_student_df = pd.DataFrame(index=range(num_students), columns=student_df.columns)
    
    # 生成各种数据列表
    # 学号
    stu_no = [f"2024{i:03d}" for i in range(1, num_students + 1)]
    
    # 姓名
    surnames = ['李', '王', '张', '刘', '陈', '杨', '赵', '黄', '周', '吴',
                '徐', '孙', '胡', '朱', '高', '林', '何', '郭', '马', '罗']
    given_names = ['伟', '芳', '娜', '秀', '敏', '静', '丽', '强', '磊', '军',
                   '洋', '勇', '艳', '杰', '娟', '涛', '明', '超', '华', '美']
    
    stu_names = []
    used_names = set()
    while len(stu_names) < num_students:
        surname = random.choice(surnames)
        if random.random() < 0.3:
            given_name = random.choice(given_names)
        else:
            given_name = random.choice(given_names) + random.choice(given_names)
        full_name = surname + given_name
        if full_name not in used_names:
            used_names.add(full_name)
            stu_names.append(full_name)
    
    # 性别
    stu_gender = ['男'] * 350 + ['女'] * 350
    random.shuffle(stu_gender)
    
    # 出生日期（高三学生，大约2005-2007年出生）
    stu_birthdays = []
    start_date = datetime(2005, 1, 1)
    end_date = datetime(2007, 12, 31)
    for _ in range(num_students):
        random_days = random.randint(0, (end_date - start_date).days)
        birthday = start_date + timedelta(days=random_days)
        stu_birthdays.append(birthday.strftime("%Y-%m-%d"))
    
    # 年级
    stu_grade = ['高三'] * num_students
    
    # 班级
    stu_class = [f"{random.randint(1, 20)}班" for _ in range(num_students)]
    
    # 在校状态
    stu_status = ['在读'] * num_students
    
    # 身份证号（确保不重复）
    stu_id_numbers = []
    used_ids = set()
    while len(stu_id_numbers) < num_students:
        id_number = ''.join(random.choices(string.digits, k=18))
        if id_number not in used_ids:
            used_ids.add(id_number)
            stu_id_numbers.append(id_number)
    
    # 学籍号
    stu_xuejihao = [f"2024{str(i).zfill(6)}" for i in range(1, num_students + 1)]
    
    # 入学日期
    stu_enrollment_date = ['2024-09-01'] * num_students
    
    # 使用列名将列表数据添加到对应的列中
    new_student_df['学号 (必填)'] = stu_no
    new_student_df['姓名 (必填)'] = stu_names
    new_student_df['性别 (男/女)'] = stu_gender
    new_student_df['出生日期 (YYYY-MM-DD)'] = stu_birthdays
    new_student_df['年级 (初一/初二/初三/高一/高二/高三)'] = stu_grade
    new_student_df['班级名称 (1班-20班)'] = stu_class
    new_student_df['在校状态 (在读/转学/休学/复学/毕业)'] = stu_status
    new_student_df['身份证号码'] = stu_id_numbers
    new_student_df['学籍号'] = stu_xuejihao
    new_student_df['入学日期 (YYYY-MM-DD)'] = stu_enrollment_date
    
    # 其他列保持为空（NaN）
    
    return new_student_df

# 生成正确的学生信息
student_info = generate_correct_stu_info()
# 保存到新的Excel文件
output_filename = 'generated_students.xlsx'
student_info.to_excel(output_filename, index=False)
```

```python
# 生成学生考试信息函数
def generate_student_grades():
    # 读取生成的学生信息文件
    try:
        student_df = pd.read_excel('generated_students.xlsx')
        print(f"成功读取学生信息，共{len(student_df)}条记录")
        print("列名：", student_df.columns.tolist())
    except FileNotFoundError:
        print("错误：找不到 'generated_students.xlsx' 文件")
        return None
    except Exception as e:
        print(f"读取文件时出错：{e}")
        return None
    
    # 定义科目和成绩范围
    subjects = ['语文', '数学', '英语', '政治', '历史', '物理', '化学', '生物', '地理', '体育', '美术', '音乐', '心理', '信息技术']
    
    # 创建结果列表
    result_data = []
    
    # 为每个学生生成考试成绩
    for index, row in student_df.iterrows():
        # 创建学生成绩记录
        student_record = {
            '学号': row['学号 (必填)'],  # 使用正确的列名
            '学生姓名': row['姓名 (必填)']  # 使用正确的列名
        }
        
        # 为每个科目生成成绩
        for subject in subjects:
            if subject in ['语文', '数学', '英语']:
                grade = random.randint(60, 150)  # 语数英：60-150分
            elif subject == '体育':
                grade = random.randint(40, 60)   # 体育：40-60分
            else:
                grade = random.randint(50, 100)  # 其他科目：50-100分
            
            student_record[subject] = grade
        
        result_data.append(student_record)
    
    # 创建DataFrame
    grades_df = pd.DataFrame(result_data)
    
    return grades_df

student_grades_fixed = generate_student_grades()
grades_filename = 'student_grades_1.xlsx'
student_grades_fixed.to_excel(grades_filename, index=False)
```
